```python
from bs4 import BeautifulSoup
import urllib3
import requests
import time
import winsound
duration = 600  # milliseconds
freq = 800  # Hz
```

```python
def get_data(anime):
    url = "https://myanimelist.net/anime/"+anime+"/userrecs"
    response = requests.get(url)
    if response.status_code <200 or response.status_code >299:
        print("Retrying in 3s...")
        time.sleep(3)
        response = requests.get(url)
        if response.status_code < 200 or response.status_code > 299:
            print("Somethings wrong",response.status_code, "on "+anime)
            return ['','','','',''],['','','','',''],['','','','',''],['','','','','']

    html = response.content
    bs = BeautifulSoup(html, features="html.parser")
    
    art_count=0
    sound_count=0
    vibe_count=0
    humor_count=0
    
    
    art_array=[]
    sound_array=[]
    vibe_array=[]
    humor_array=[]
    
    
    divs = bs.body.findAll("div", class_="borderClass")
    if divs is None:
        return ['','','','',''],['','','','',''],['','','','',''],['','','','','']
    j=0
    i=0
    
    while i < len(divs):
        div=divs[i]
        a=div.find("a", class_="hoverinfo_trigger")
        rectextt=div.find("div", class_="borderClass bgColor1")
        if (rectextt is None):
            return ['','','','',''],['','','','',''],['','','','',''],['','','','','']
            
        rectext=rectextt.getText()
        rec=rectext.lower()
        i=i+1

    
        recs=div.find("div", id=("simaid"+str(j)))
        if(not(recs is None)):
            recbg1=recs.findAll("div", class_="borderClass bgColor1")
            recbg2=recs.findAll("div", class_="borderClass bgColor2")
            for r in recbg1:
                i=i+1
                rec=rec+r.getText().lower()
        
            for r in recbg2:
                i=i+1
                rec=rec+r.getText().lower()
            
    
        art=0
        sound=0
        vibe=0
        humor=0
    
        if (" art" in rec or "art style" in rec or "artstyle" in rec or " arts" in rec or "art styles" in rec or "artstyles" in rec or "animation" in rec or "animations" in rec) and art_count<5:
            art = 1
            art_count+=1
            art_array.append(list(a['href'].split("/"))[4])

        if ("soundtrack" in rec or "soundtracks" in rec or "music" in rec or " ost" in rec or " osts" in rec or "songs" in rec) and sound_count<5:
            sound=1
            sound_count+=1
            sound_array.append(list(a['href'].split("/"))[4])

            
        if (" world" in rec or " worlds" in rec or "setting" in rec or "settings" in rec or "vibe" in rec or "vibes" in rec or " feel" in rec or "background" in rec or "backgrounds" in rec 
            or "themes" in rec or "theme" in rec or "plot" in rec or "plots" in rec or "protagonists" in rec) and vibe_count<5:
            vibe=1
            vibe_count+=1
            vibe_array.append(list(a['href'].split("/"))[4])

        if ("funny" in rec or "humor" in rec or "hilarious" in rec or "comedy" in rec or "humorous" in rec or "humour" in rec or "humourous" in rec) and humor_count<5:
            humor=1
            humor_count+=1
            humor_array.append(list(a['href'].split("/"))[4])
        
        i=i+1
        j=j+1
    
    
    while len(art_array)<5:
        art_array.append("")

    while len(sound_array)<5:
        sound_array.append("")

    while len(vibe_array)<5:
        vibe_array.append("")
        
    while len(humor_array)<5:
        humor_array.append("")
    
    
    return art_array,sound_array,vibe_array,humor_array
```

```python
import csv
import pandas as pd

with open('anime.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    
    df1=pd.read_csv('archive/blank.csv',names=['MAL_ID', 'Name', 'Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5'], dtype={'MAL_ID': 'str', 'Name': 'str', 'Anime-1': 'str', 'Anime-2': 'str', 'Anime-3': 'str', 'Anime-4': 'str','Anime-5': 'str'})
    df2=pd.read_csv('archive/blank.csv',names=['MAL_ID', 'Name', 'Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5'], dtype={'MAL_ID': 'str', 'Name': 'str', 'Anime-1': 'str', 'Anime-2': 'str', 'Anime-3': 'str', 'Anime-4': 'str','Anime-5': 'str'})
    df3=pd.read_csv('archive/blank.csv',names=['MAL_ID', 'Name', 'Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5'], dtype={'MAL_ID': 'str', 'Name': 'str', 'Anime-1': 'str', 'Anime-2': 'str', 'Anime-3': 'str', 'Anime-4': 'str','Anime-5': 'str'})
    df4=pd.read_csv('archive/blank.csv',names=['MAL_ID', 'Name', 'Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5'], dtype={'MAL_ID': 'str', 'Name': 'str', 'Anime-1': 'str', 'Anime-2': 'str', 'Anime-3': 'str', 'Anime-4': 'str','Anime-5': 'str'})

    i=0
    for row in reader:
        if(row[0]=='MAL_ID'):
            continue
        anime=row[0]+"/"+row[1].replace(" ","_")
        art_array,sound_array,vibe_array,humor_array=get_data(anime)
        time.sleep(2)

 
        

        df1.loc[df1['MAL_ID']==row[0],['Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5']]=[art_array[0],art_array[1],art_array[2],art_array[3],art_array[4]]
        df2.loc[df2['MAL_ID']==row[0],['Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5']]=[sound_array[0],sound_array[1],sound_array[2],sound_array[3],sound_array[4]]
        df3.loc[df3['MAL_ID']==row[0],['Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5']]=[vibe_array[0],vibe_array[1],vibe_array[2],vibe_array[3],vibe_array[4]]
        df4.loc[df4['MAL_ID']==row[0],['Anime-1', 'Anime-2', 'Anime-3', 'Anime-4', 'Anime-5']]=[humor_array[0],humor_array[1],humor_array[2],humor_array[3],humor_array[4]]
        
        i=i+1
        
        
        if i%10==0:
            print("Completed "+str(i)+" animes")
            print("At "+row[0]+" anime")
            

        if i%100==0:
            df1.to_csv("archive/art1.csv", index=False, header=False )
            df2.to_csv("archive/sound1.csv", index=False, header=False )
            df3.to_csv("archive/vibe1.csv", index=False, header=False )
            df4.to_csv("archive/humor1.csv", index=False, header=False )
        
        
        
```
